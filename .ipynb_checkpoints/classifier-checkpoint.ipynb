```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sbn

#Sk-learn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder , OneHotEncoder, LabelBinarizer, normalize
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, multilabel_confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
#Build and train models:

Y_Train= [] 
Y_Test = [] 

def one_hot(y, nclass = 0):
  if y.ndim > 1 : 
    vec = np.zeros(y.shape)
    idx = np.argmax(y, axis=1)
    vec[np.arange(y.shape[0]), idx] =1
  else:
    nclass = np.max(np.max(y)+1 , nclass)
    vec    = np.zeros((y.shape[0] , nclass))
    vec[np.arange(y.shape[0]), y] = 1
  return vec

def append_it(y1, y2):
  Y_Train.append(y1)
  Y_Test.append(y2)
  
def save_data():
  Y1 = np.transpose(np.asarray(Y_Train))
  Y2 = np.transpose(np.asarray(Y_Test))
  np.savetxt('train.txt', Y1 , '%4d','\t')
  np.savetxt('test.txt', Y2 , '%4d','\t')
  
def print_results(mname, model,X_train, X_test, y_train, y_test ):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    y_train_onehot = one_hot(y_train)
    y_test_onehot  = one_hot(y_test)
    y_train_pred_onehot = one_hot(y_train_pred)
    y_test_pred_onehot = one_hot(y_test_pred)
    
    print('-'*50 ,'\n')
    print('Model:\t' , mname,
          'Train acc:\t', accuracy_score(y_train_onehot, y_train_pred_onehot), 
          '\tTest acc:\t', accuracy_score(y_test_onehot, y_test_pred_onehot))
    print('Train report')
    print(classification_report(y_train_pred_onehot, y_train_onehot))
    print('Test report')
    print(classification_report(y_test_pred_onehot, y_test_onehot))
    append_it(y_train_pred, y_test_pred)
  

def run_models(X_train, X_test, y_train, y_test):
    #make a calss label copy of the one-hot labels
    y_train_labels = np.argmax(y_train, axis=1) 
    y_test_labels = np.argmax(y_test, axis=1) 
  
    
    # ___________________ using KNN__________________#
    model_name = 'KNN'
    print(model_name)
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(X_train,y_train)
    print_results(model_name, model,X_train, X_test, y_train, y_test )
  

    # ___________________ Neural Networks - Classic _______________#
    model_name = 'Neural Networks - Classic'
    print(model_name)
    model = MLPClassifier()
    MLPClassifier(solver='adam', activation='tanh', early_stopping=True, alpha=1e-5,
                  batch_size=4, hidden_layer_sizes=(50, 20, 5), random_state=1)
    model.fit(X_train, y_train)
    print_results(model_name, model,X_train, X_test, y_train, y_test )

    # ___________________ Logistic Regression__________________#
    model_name = 'Logistic Regression'
    print(model_name)
    model = LogisticRegression()
    model = model.fit(X=X_train, y=y_train_labels)
    print_results(model_name, model,X_train, X_test, y_train_labels, y_test_labels )

    # ___________________ using naive bayes__________________#
    model_name = 'Naive Bayes'
    print(model_name)
    model = GaussianNB()
    model = model.fit(X=X_train, y=y_train_labels)
    print_results(model_name, model,X_train, X_test,  y_train_labels, y_test_labels )

    # _____________________Soppurt vector machine________________#
    model_name = 'SVM-RBF'
    print(model_name)
    model = svm.SVC(kernel='rbf', gamma='auto', C=100)
    model = model.fit(X=X_train, y=y_train_labels)
    print_results(model_name, model,X_train, X_test,  y_train_labels, y_test_labels )

    # ___________________________ Random forest___________________#
    model_name = 'Random Forest'
    print(model_name)
    model = RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train_labels)
    print_results(model_name, model,X_train, X_test,  y_train_labels, y_test_labels )

    # ______________Decision Tree________________#
    model_name = 'Decisoin Tree'
    print(model_name)
    model = tree.DecisionTreeClassifier()
    model.fit(X_train, y_train_labels)
    print_results(model_name, model,X_train, X_test, y_train_labels, y_test_labels )

    
    # __________Adaboots classifier__________#
    model_name = 'Adaboost'
    print(model_name)
    model = AdaBoostClassifier(n_estimators=500)
    model.fit(X_train, y_train_labels)
    print_results(model_name, model,X_train, X_test, y_train_labels, y_test_labels )

    # __________Ensemble __________#
    model_name = 'Ensemble of the all'
    print(model_name)
    estimators = []
    model1 = LogisticRegression()
    estimators.append(('logistic', model1))
    model2 = GaussianNB()
    estimators.append(('nb', model2))
    model3 = AdaBoostClassifier(n_estimators=100)
    estimators.append(('adaboost', model3))
    model4 = MLPClassifier()
    MLPClassifier(solver='adam', activation='tanh', early_stopping=True, alpha=1e-5,
                  batch_size=64, hidden_layer_sizes=(100, 50, 10), random_state=1)
    estimators.append(('nn', model4))
    model5=RandomForestClassifier(n_estimators=100)
    estimators.append(('rf', model5))
    # create the ensemble model
    model = VotingClassifier(estimators)
    model.fit(X_train, y_train_labels)
    print_results(model_name, model,X_train, X_test, y_train_labels, y_test_labels )
     
         
  
```

```python
class CarClass:

  def __init__(self, dfs):
    self.dataframe = dfs
    self.dataset   = self.extractFeatures()
    self.target    = dfs["label"]
    self.classes   = ["suv", "sedan", "bus", "microbus", "truck", "minivan"]
    self.features  = ["width", "length", "diameter", "ratio", "perimeter", "area"]

  def extractFeatures(self):
    width     = abs(self.dataframe["x2"] - self.dataframe["x1"])
    length    = abs(self.dataframe["y2"] - self.dataframe["y1"])
    perimeter = 2*(width + length)
    ratio     = np.int8(length / width)
    area      = width * length
    diameter  = np.sqrt(width**2 + length**2)
    dataset   = np.array((width, length, diameter, ratio, area, perimeter)).transpose()
    return dataset
```

```python

file_path = "cars-labeled.csv"
dfs       = pd.read_csv(file_path)

car_dataset   = CarClass(dfs)
dataset_train, dataset_test, target_train, target_test = train_test_split(car_dataset.dataset, 
                                                                          car_dataset.target, 
                                                                          random_state=1) 


```

```python

clf = DecisionTreeClassifier(max_depth = 3, random_state = 0)
clf.fit(dataset_train, target_train)
clf.predict(dataset_test)

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)

tree.plot_tree(clf,
               feature_names = car_dataset.features, 
               class_names=car_dataset.classes,
               filled = True);

```

```python
rf = RandomForestClassifier(n_estimators=100,
                            random_state=0) # criterion = "gini"
rf.fit(dataset_train, target_train)
predicts = rf.predict(dataset_test)

# predicts_lb    = LabelBinarizer().fit(predicts).transform(predicts)
# target_test_lb = LabelBinarizer().fit(target_test).transform(target_test)

CM = confusion_matrix(target_test, predicts, normalize="true")
print("Confusion Matrix: \n", CM)

sbn.heatmap(CM, annot=True, fmt=".2f")
plt.show()
```

```python
clf = MLPClassifier(solver='adam', activation='tanh', early_stopping=True, alpha=1e-5,
                  batch_size=4, hidden_layer_sizes=(50, 20, 5), random_state=0).fit(dataset_train, target_train)
predicts = clf.predict(dataset_test)

CM = confusion_matrix(target_test.values, predicts, normalize="true")
sbn.heatmap(CM, annot=True)

print(predicts)
print(target_test.values)
```

```python
## Run tests
#Load iris data set and add a column of 1's to features

from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
X = np.c_[np.ones(X.shape[0]) , X]

X_train, X_test, y_train, y_test = train_test_split(X, one_hot(y), test_size=0.20, random_state=42)

print ('Train size:', X_train.shape, '\t' , y_train.shape)
print ('Test size:',  X_test.shape,  '\t' , y_test.shape)

run_models(X_train , X_test , y_train , y_test)


```
